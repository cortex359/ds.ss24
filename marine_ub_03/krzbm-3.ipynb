(9) \[Optional\] Machen Sie Ihre Abbildung interaktiv. Nutzen Sie die `interact` Funktion, um eine Abbildung zu erzeugen, in der Sie mithilfe eines Reglers das Jahr einstellen können, für das die Abbildung erzeugt werden soll. Wie verändern sich Lebenserwartung und GDP per Capita über die Jahre hinweg? (1-3 Sätze)

* Hinweis: Um diese Teilaufgabe umzusetzen, recherchieren Sie im Netz und lesen Sie unter anderem [hier](https://ipywidgets.readthedocs.io/en/stable/examples/Using%20Interact.html) nach, wie Sie `interact` nutzen können.

```python
from ipywidgets import interact
import pandas as pd
import matplotlib.pyplot as plt

def plot_scatter(year):
    # Filtere den Datensatz für das ausgewählte Jahr
    data_year = gapminder[gapminder['year'] == year]
    
    # Definiere die Achsen
    x = data_year['gdpPercap']
    y = data_year['lifeExp']
    sizes = data_year['pop'] / 1000000  # Skaliere die Größe der Punkte entsprechend der Population

    # Identifiziere die Kontinente im Datensatz
    continents = data_year['continent'].unique()

    # Erstelle eine Farbpalette für die Kontinente
    colors = plt.cm.tab10.colors[:len(continents)]

    # Erstelle den Scatter-Plot für jeden Kontinent
    plt.figure(figsize=(10, 6))
    for i, continent in enumerate(continents):
        continent_data = data_year[data_year['continent'] == continent]
        plt.scatter(continent_data['gdpPercap'], continent_data['lifeExp'], s=sizes[continent_data.index], alpha=0.5, label=continent, color=colors[i])

    # Achsenbeschriftungen und Titel
    plt.xscale('log')  # Logarithmische Skalierung der x-Achse
    plt.xlabel('GDP per Capita')
    plt.ylabel('Life Expectancy')
    plt.title(f'Life Expectancy vs GDP per Capita (Year: {year})')

    # Legende anzeigen
    plt.legend(title='Continent')

    # Zeige den Plot
    plt.grid(True)
    plt.show()

# Interaktive Erzeugung der Abbildung mit einem Regler für das Jahr
interact(plot_scatter, year=(gapminder['year'].min(), gapminder['year'].max(), 1))
```

### 1 Hypothesen "Total Fertility"

#### 1.0 Vorbereitung

```python
total_fertility = pd.read_csv('children_per_woman_total_fertility.csv')
total_fertility_per_country = total_fertility.set_index('country')
print(total_fertility_per_country.head())
today = datetime.date.today()
year = today.year

years_until_now = total_fertility_per_country.columns[total_fertility_per_country.columns <= str(year)]
total_fertility_per_country_until_now = total_fertility_per_country[years_until_now]
```

```python
def first_quartile(fertility):
    return fertility.quantile(0.25)

def third_quartile(fertility):
    return fertility.quantile(0.75)

fertility_subtractions = pd.DataFrame()

fertility_subtractions["Werte bis 1952:"] = total_fertility_per_country_until_now["1952"] - total_fertility_per_country_until_now[years_until_now[0]]


fertility_subtractions["Werte ab 1953:"] = total_fertility_per_country_until_now[years_until_now[len(years_until_now) - 1]] - total_fertility_per_country_until_now["1953"]

fertility_subtractions["Werte bis 1952 pro 10 Jahre:"] = (total_fertility_per_country_until_now["1952"] - total_fertility_per_country_until_now[years_until_now[0]]) / (1952 - int(years_until_now[0])) * 10

fertility_subtractions["Werte ab 1953 pro 10 Jahre:"] = (total_fertility_per_country_until_now[years_until_now[len(years_until_now) - 1]] - total_fertility_per_country_until_now["1953"]) / (int(years_until_now[len(years_until_now) - 1]) - 1953) * 10

print("Absolute Werte:")
print(fertility_subtractions[["Werte bis 1952:", "Werte ab 1953:"]].agg(["mean", "std", "min", "max", "median", first_quartile, third_quartile]))
print("_________________________________________________")
print("\nRelative Werte:")
print(fertility_subtractions[["Werte bis 1952 pro 10 Jahre:", "Werte ab 1953 pro 10 Jahre:"]].agg(["mean", "std", "min", "max", "median", first_quartile, third_quartile]))
```

##### Auswertung:
Anhand des maximalen Wertes und dem 3. Quartil der Werte bis 1952, kann man erkennen, dass tatsächlich in einigen Ländern die Fertilität bis 1952 gestiegen sind (mindestens in 25 % der Länder).  
Allerdings kann man anhand der Durchschnittswerte auch beobachten, dass der Großteil der Länder trotzdem einen leichten fall der Fertilität zu verzeichnen hat.  
Wenn man aber die Werte bis 1952 mit denen ab 1953 vergleicht, merkt man ganz klar, dass der Abfall ab 1952 stark angestiegen ist.  
Besonders, wenn man sich die relativen Werte ansieht. Nach 1952 hat sich der durchschnittliche Abfall mehr als verzehnfacht.

#### 1.4 Die Anzahl der Länder mit fallender Fertilität steigt kontinuierlich und nicht plötzlich

```python
countries_with_subtraction = total_fertility_per_country_until_now.copy()
for i in range(len(years_until_now) -1):
    countries_with_subtraction[years_until_now[i+1]] -= total_fertility_per_country_until_now[years_until_now[i]]

number_of_countries_with_negative_fertilitydiff = (countries_with_subtraction < 0).sum()
number_of_countries_with_negative_fertilitydiff.plot(kind="line", xlabel='years', ylabel='# Countries with decreasing fertility')
derivative_of_number_of_countries_with_negative_fertilitydiff = number_of_countries_with_negative_fertilitydiff.copy()
for i in range(len(years_until_now) -1):
    derivative_of_number_of_countries_with_negative_fertilitydiff[years_until_now[i+1]] -= number_of_countries_with_negative_fertilitydiff[years_until_now[i]]
plt.show()
derivative_of_number_of_countries_with_negative_fertilitydiff.plot(kind="line", xlabel='years', ylabel='Diff Countries with decreasing fertility')
plt.show()
```

##### Auswertung:
Hier kann man beobachten, dass die Anzahl der Länder mit sich verringender Fertilität zwar makroskopisch betrachtet gestiegen ist, aber auch dort dies mit kleineren Peaks und regelmäßigen Tälern passiert.  
Zusätzlich ist interessant zu sehen, dass der größte Anstieg (Zwischen 1955 und 1990) in einem stattfand, sowie, dass nach einem Peak, die Anzahl wieder abgefallen ist.  
Dies müsste man mit der Anzahl der steigende und der Stagnierenden Länder vergleichen, um zu beobachten, ob sich in diesen Tälern die Fertilität erholt oder nicht

#### 1.4.1 Bei sinkender Anzahl von Ländern, deren Fertilität sinkt, erholt sich die Fertilität nicht, sondern stagniert 

```python
fertility_differences_amount = pd.DataFrame()
fertility_differences_amount["Decreasing"] = (countries_with_subtraction < 0).sum()
fertility_differences_amount["Stagnating"] = (countries_with_subtraction == 0).sum()
fertility_differences_amount["Increasing"] = (countries_with_subtraction > 0).sum()

fertility_differences_amount.plot(kind='line', xlabel='Years', ylabel='Amount of Countries')
plt.show()
```

##### Auswertung:
Aus der Grafik kann man erkennen, dass wenn die Anzahl der Länder mit sich verringernder Fertilität sinkt, die Anzahl der Länder mit sich erhöhender Fertilität steigt.  
Somit kann man davon ausgehen, dass in den zuvor genannten Tälern, sich die Fertilität leicht erholt

#### 1.6 Die tatsächliche Veränderung der Fertilität ist wichtiger, als die Anzahl der Länder mit verändernder Fertilität pro Jahr 

```python
sumation_negativity_per_year = countries_with_subtraction[countries_with_subtraction < 0].fillna(0).sum()
sumation_negativity_per_year_per_country = sumation_negativity_per_year / number_of_countries_with_negative_fertilitydiff
print(sumation_negativity_per_year_per_country.agg(['mean', 'std', 'min', 'max'], axis=0))
sumation_negativity_per_year_per_country.plot(kind='line')
plt.show()
(-1 * number_of_countries_with_negative_fertilitydiff).plot(kind='line')
plt.show()
sumation_negativity_per_year.plot(kind='line')
plt.show()
mean_fertility_per_year = total_fertility_per_country_until_now.mean()
mean_fertility_differentiated = mean_fertility_per_year.copy()
mean_fertility_differentiated[years_until_now[0]] -= mean_fertility_differentiated[years_until_now[0]]
for i in range(len(years_until_now) -1):
    mean_fertility_differentiated[years_until_now[i+1]] -= mean_fertility_per_year[years_until_now[i]]

mean_fertility_differentiated.plot(kind='line')
plt.show()
```
